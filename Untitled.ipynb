```python
# Define callback function to handle received messages
def on_message(message):
    try:
        # Ensure the message has 'data' which is expected to be a list of quotes
        if isinstance(message, dict) and 'data' in message:
            # Loop through each entry in the 'data' array
            for data_point in message['data']:
                # Extract instrument token and last price (ltp) if they exist
                instrument_token = data_point.get("tk")
                last_price = data_point.get("ltp")
                
                if instrument_token and last_price is not None:
                    # Display formatted information for each token
                    print(f"Instrument Token: {instrument_token} | Last Traded Price (LTP): {last_price}")
                else:
                    # If required fields are missing, display the received data for debugging
                    print("Incomplete data received:", data_point)
        else:
            print("Received message in an unexpected format:", message)
    except Exception as e:
        print("Error processing message:", e)

# Set up the client to use the defined on_message function
client.on_message = on_message

# Define instrument tokens for subscription
inst_tokens = [
    {"instrument_token": "50393", "exchange_segment": "nse_fo"},
    {"instrument_token": "46144", "exchange_segment": "nse_fo"},
    {"instrument_token": "46146", "exchange_segment": "nse_fo"},
    {"instrument_token": "46150", "exchange_segment": "nse_fo"},
    {"instrument_token": "46152", "exchange_segment": "nse_fo"},
    {"instrument_token": "46156", "exchange_segment": "nse_fo"},
]

# Subscribe to tokens to receive live LTP data
client.subscribe(instrument_tokens=inst_tokens, isIndex=False, isDepth=False)

# Keep the connection alive in Jupyter Notebook
import time
while True:
    time.sleep(1)
```

```python
def get_option_chain(symbol, strike_price, expiry=""):
    # Search all options related to the specified symbol, strike price, and expiry date
    search_response = client.search_scrip(
        exchange_segment="nse_fo",  # NSE Futures and Options segment
        symbol=symbol,
        strike_price=str(strike_price),
        option_type="PE",           # 'PE' for Put, 'CE' for Call
        expiry=expiry               # Use a specific expiry date or leave it blank
    )
    
    # Process and display the retrieved option chain
    if search_response:
        print(f"Option chain for {symbol} at strike price {strike_price}:\n")
        for scrip in search_response:
            print(
                f"Instrument Token: {scrip['pSymbol']}, "
                f"Trading Symbol: {scrip['pTrdSymbol']}, "
                f"Expiry Date: {scrip['pExpiryDate']}, "
                f"Option Type: {scrip['pInstType']}"
            )
    else:
        print("No options found for the given criteria.")

strike_prices = range(23900, 24600, 50)  # Adjust the range and step as needed

for strike_price in strike_prices:
    get_option_chain(symbol="NIFTY", strike_price=strike_price, expiry="08nov2024")
```
